```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
```

```python
frame = 1024;
shift = frame//4;
filename = 's01_4ch'
nch = 4
```

```python
# [raw_matlab,fs1] = audioread('../input/F01_22GC010A_BUS.wav');
[raw_python,fs2] = sf.read('C:/git_repository/2024-LG-CTO-Python-Integrated-System2/' + filename +'_out.wav');
[raw_cpp,fs2] = sf.read('../output/'  + filename + '_output.wav');
```

```python
plt.figure()
for ch in range(nch):
    plt.subplot(nch,3,ch*3+1)
    plt.plot(raw_python[:,ch])
    plt.subplot(nch,3,ch*3+2)
    plt.plot(raw_cpp[:,ch])
    tmpc = raw_cpp[frame-shift:,ch]
    tmpp = raw_python[:len(tmpc),ch]
    plt.subplot(nch,3,ch*3+3)
    plt.plot(tmpc-tmpp)
    plt.show()
```
